# Results interpretation

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
csv_filepath = '../results/results_EuCoBase.csv'
```

```python
df = pd.read_csv(csv_filepath)
df.head(5)
```

```python
# Generate loss curve
plt.plot(df['epoch'], df['train_loss_bpr'], label='train')
plt.plot(df['epoch'], df['test_loss_bpr'], label='test')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```

```python
# Generate evaluation metrics plot
plt.plot(df['epoch'], df['precision@k'], label='precision@k')
plt.plot(df['epoch'], df['recall@k'], label='recall@k')
plt.plot(df['epoch'], df['map@k'], label='map@k')
plt.plot(df['epoch'], df['mrr@k'], label='mrr@k')
plt.plot(df['epoch'], df['ndcg@k'], label='ndcg@k')
plt.plot(df['epoch'], df['hit_rate@k'], label='hit_rate@k')
plt.xlabel('Epoch')
plt.ylabel('Performance')
plt.title('Evaluation metrics')
plt.legend()
plt.show()
```
